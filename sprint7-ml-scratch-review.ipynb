#  RecallとPrecisionはどう違うか

Recall:実際に陽性だったもののうち、モデルが陽性と判断したものの割合を示す。
この指標は、怪しいものはとりあえず全て見付け出したい時に重視すべきもので、この評価指標を重視するとわかりにくいものは基本的に陽性と判断することになるので、その中には実際には陰性だったものがたくさん含まれる。
つまり再現率を重視するときはFPが発生することを許容できるケースと言える。

Precison:適合率
モデルが陽性と判断したものの中に、どれだけ陽性なものが含まれていたかを判断する
いわば、正確性を見るための指標と言える。
この指標は明らかに陽性とわかりやすいものを見つけたい時に重視するもの。
この評価指標を重視すると、わかりにくいものが基本的に陰性と判断することになるので、その中には実際には陽性だったものがたくさん含まれる。


たくさん陽性を出す検査（保守的な検査）は再現率は高くなるが、適合率は低くなる。逆に、たくさん陰性を出す検査は適合率は高くなるが、再現率は低くなる。つまり、再現率と適応度にはトレードオフの関係がある。

調査の目的によって、例えば病院の診断では、実際は病気なのに病気でないと診断してしまうことを防ぐことに注力することが診断の性質上もっとも求められる。つまり、  FNを減らすことが評価指標を作成する際に要請されることなので、再現率が用いられる。一方でスパムメールの判定のように、重要なメールをスパムと判定し、見れなくなってしまうことを防がなければならない場合では適合率が用いられる。

# 過学習を防ぐための方法を3つ以上説明する

ミニバッチ処理:サンプルのデータをいくつかに分けて、ランダムに選んで学習させる。  
L1正則化(ラッソ回帰)  :L1正則化項によって、不要なパラメータを削ることができる。予測したいデータに対する寄与が薄いデータ」や「他の予測に用いられるデータとの関係性が強いデータ」の正則化に向いており、これらを回帰分析の際に結果に対する寄与が小さくなるように係数を小さくする方法。 データとして余分な情報がたくさん存在するようなデータ の回帰分析を行う際に重宝する。
 データとして余分な情報がたくさん存在するようなデータの回帰分析を行う際 に重宝すると確認しました。そのため、データセットの数（行数）に比べて、パラメータの数（列数）が多いなどといった場合には、ラッソ回帰を利用する.  
L2正則化(リッジ回帰) : 予測に用いるデータの範囲を算出し、データの範囲を揃えるようにデータに対する係数を小さくすることによって回帰分析のモデルの一般化を図ろうとする方法。 データの範囲とは、データが取りうる数値の範囲のことで、揃える場合は大抵は0から1の範囲になるように調整される。データの範囲を揃えることによって同じ尺度で全てのデータの予測に対する寄与が比較可能になり、 滑らかなモデルを得やすい（汎化しやすい）

# バイアスとバリアンスのトレードオフとは  
バイアスとは実際の値と予測値の誤差の平均のことで、バイリアンスは予測値がどれだけ散らばっているかを示す。
つまりバイアスが高い状態は、機械学習モデルはしっかり学習をできていない可能性が高く、予測値の制度が悪い.バリアンスが高い状態モデルが過学習している。
バイアスとバリアンスのトレードオフとは、低バイアスの状態（予測値と実際値の誤差が少ない）になるが、その一方でバリアンスは高まり過学習に陥ること。

また決定木は別の特徴量との関係性は見られないので前処理として、スケーリングしたり、欠損値の処理をする必要はない。

# これ以外にどのような機械学習手法があるか簡潔にまとめる  


 ElasticNet回帰:ラッソ回帰とリッジ回帰を組み合わせて正則化項を作るモデル.  メリットとしては、ラッソ回帰で取り扱った 余分な情報がたくさん存在するようなデータに対して情報を取捨選択してくれる点 と、リッジ回帰で取り扱った 滑らかなモデルを得やすい（汎化しやすい）点 の組み合わせとなるので、両方のメリットをバランスよく用いてモデルを作りたい時にはベストな手法.
